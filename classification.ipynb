## Preprocessing

```python
df.isnull().sum()
```

```python
print("Total Rows :",len(df.index))
```

**Deleting the columns** that have large portion of values `null`.

```python
df = df.drop(['Source of Payment 3', 'Other Provider License Number'], axis = 1)
```

```python
df.isnull().sum()
```

Removing rest of the `null` values containing rows.

```python
modified_df = df.dropna(subset=['Health Service Area', 'Hospital County', 'Operating Certificate Number', 'Facility ID', 'Zip Code - 3 digits', 'Patient Disposition', 'CCS Diagnosis Code', 'CCS Diagnosis Description', 'CCS Procedure Code', 'CCS Procedure Description', 'APR Severity of Illness Description', 'APR Risk of Mortality', 'Attending Provider License Number'])
```

```python
modified_df.isnull().sum()
```

**Placing mode on missing value** on the ramaining `null` rows.

```python
variables =['Source of Payment 2','Operating Provider License Number']
modified_df[variables] = modified_df[variables].fillna(modified_df[variables].mode().iloc[0])
```

```python
modified_df.isnull().sum()
```

### Label Encoding

```python
modified_df.info()
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()

columns = ['Health Service Area', 'Hospital County', 'Facility Name', 'Age Group', 'Zip Code - 3 digits', 'Gender', 'Race', 'Ethnicity', 'Length of Stay', 'Type of Admission', 'Patient Disposition', 'CCS Diagnosis Description', 'CCS Procedure Description', 'APR DRG Description', 'APR MDC Description', 'APR Severity of Illness Description', 'APR Risk of Mortality', 'APR Medical Surgical Description', 'Source of Payment 1', 'Source of Payment 2', 'Abortion Edit Indicator', 'Emergency Department Indicator']

for column in columns:
    modified_df[column + ' Encoded'] = le.fit_transform(modified_df[column])
```

## Classification of Risk of Mortality
The risk of mortality provides a medical classification to estimate the likelihood of in-hospital death for a patient.

```python
X = [
 'Operating Certificate Number',
 'Facility ID',
 'Discharge Year',
 'CCS Diagnosis Code',
 'CCS Procedure Code',
 'APR DRG Code',
 'APR MDC Code',
 'APR Severity of Illness Code',
 'Attending Provider License Number',
 'Operating Provider License Number',
 'Birth Weight',
 'Total Charges',
 'Total Costs',
 'Health Service Area Encoded',
 'Hospital County Encoded',
 'Facility Name Encoded',
 'Age Group Encoded',
 'Zip Code - 3 digits Encoded',
 'Gender Encoded',
 'Race Encoded',
 'Ethnicity Encoded',
 'Length of Stay Encoded',
 'Type of Admission Encoded',
 'Patient Disposition Encoded',
 'CCS Diagnosis Description Encoded',
 'CCS Procedure Description Encoded',
 'APR DRG Description Encoded',
 'APR MDC Description Encoded',
 'APR Severity of Illness Description Encoded',
 'APR Medical Surgical Description Encoded',
 'Source of Payment 1 Encoded',
 'Source of Payment 2 Encoded',
 'Abortion Edit Indicator Encoded',
 'Emergency Department Indicator Encoded']

y = 'APR Risk of Mortality Encoded'
```

```python
modified_df['APR Risk of Mortality'].unique()
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(modified_df[X], modified_df[y], test_size=0.2, random_state=42)

# Create and configure the SGD Classifier
sgd_classifier = SGDClassifier(loss='hinge', alpha=0.001, max_iter=1000, random_state=42)

# Train the classifier on the training data
sgd_classifier.fit(X_train, y_train)

# Make predictions on the test data
y_pred = sgd_classifier.predict(X_test)

# Evaluate the classifier's performance (e.g., accuracy)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

```python
from sklearn.ensemble import RandomForestClassifier 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(modified_df[X], modified_df[y], test_size=0.25, random_state=42)

# Create and configure the RandomForest Classifier
rf_classifier = RandomForestClassifier(n_estimators = 100, n_jobs = -1) 

# Train the classifier on the training data
rf_classifier.fit(X_train, y_train)

# Make predictions on the test data
y_pred = rf_classifier.predict(X_test)

# Evaluate the classifier's performance (e.g., accuracy)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

### Hyper Parameter Tuning (Grid Search)

```python
rf_classifier.get_params()
```

```python
grid_param = {
    'n_estimators': [200],
    'criterion': ['gini', 'entropy'],
    'bootstrap': [False]
}
```

```python
from sklearn.model_selection import GridSearchCV

grid_search = GridSearchCV(rf_classifier, grid_param, cv=5, scoring=None, n_jobs=8)
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
best_model = grid_search.best_estimator_
print('----------------------RESULTS----------------------')
y_pred = rf_classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
print("Best Params:    ", best_params)
```
